```python
from itertools import chain, combinations
import random 
import collections
import numpy as np

def k_set(products, k):
    return chain.from_iterable(combinations(products, n) for n in range(1,k+1))


def greedy_approval(p, n, b, C, k):

    products = list(map(str, list(range(p))))
    costs = list(map(int, C))
    assert len(products) == len(costs)

    cost_dict = dict(zip(products, costs))

    possible_ballots = list(k_set(products, k))
    profile = random.choices(possible_ballots, k=n)
    flat_votes = sorted([p for ballot in profile for p in ballot])

    total_elected = []
    total_cost_elected = 0
    stop = False

    while not stop:

        if flat_votes == [] or total_cost_elected >= b:
            break

        prod_elected = collections.Counter(flat_votes).most_common(1)[0][0]
        prod_elected_cost = cost_dict[prod_elected]

        if prod_elected_cost + total_cost_elected <= b:
            total_elected.append(prod_elected)
            total_cost_elected = total_cost_elected + prod_elected_cost

        flat_votes = list(filter(prod_elected.__ne__, flat_votes))

    return total_elected


def greedy(profile, prod_costs):

    flat_votes = sorted([p for ballot in profile for p in ballot])

    total_elected = []
    total_cost_elected = 0
    stop = False

    while not stop:

        if flat_votes == [] or total_cost_elected >= b:
            break

        prod_elected = collections.Counter(flat_votes).most_common(1)[0][0]
        prod_elected_cost = cost_dict[prod_elected]

        if prod_elected_cost + total_cost_elected <= b:
            total_elected.append(prod_elected)
            total_cost_elected = total_cost_elected + prod_elected_cost

        flat_votes = list(filter(prod_elected.__ne__, flat_votes))

    return total_elected
```

```python
## strategy proofness 
# for a given situation P, C, b, n, k pick a manipulator (i) and check all other profiles this person could 
# present as a ballot & check the 'payoff' (amount of money spent on their picks). Compare with true, if
# any case where lying is better > not strategy proof. 
```

```python
def check_strategyproofness(profile, prod_costs, possible_ballots):
    
    strategy_proof = True
    for manipulator in range(n):
    
        manipulator_profile = profile[manipulator]
        
        real_elected = greedy(profile, prod_costs)
        real_overlap = set(real_elected) & set(manipulator_profile)
        real_outcome = sum([prod_costs[p] for p in real_overlap])
            
        for possible_ballot in possible_ballots:
            
            profile[manipulator] = possible_ballot
            elected = greedy(profile, prod_costs)
            overlap = set(elected) & set(manipulator_profile)
            outcome = sum([prod_costs[p] for p in overlap])
            
            if outcome > real_outcome:
                strategy_proof = False
                return strategy_proof
                
    return strategy_proof
```

```python
p = 10
C = np.zeros(p) # np.random.randint(0, 10, size=p)
k = 6
b = 10
n = 10

products = list(map(str, list(range(p))))
costs = list(map(int, C))
    
cost_dict = dict(zip(products, costs))
possible_ballots = list(k_set(products, k))
profile = random.choices(possible_ballots, k=n)

check_strategyproofness(profile, cost_dict, possible_ballots)
```
